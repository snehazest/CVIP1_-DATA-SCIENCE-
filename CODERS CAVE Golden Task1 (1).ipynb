#lets build the model to predict house price

```python
# Separate features and target variable
columns_to_keep = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
                   'waterfront', 'view', 'grade', 'yr_built', 'yr_renovated',
                   'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']
features = df[columns_to_keep]
#these feature will impact the price of the house
```

```python
#Therefore we have
features = df[columns_to_keep]
target = df['price']
```

```python
x=features
y=target
```

```python
# Split the training data into training and testing 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Dimension of dataset
x.shape,x_train.shape,x_test.shape
```

```python
#Train the model
model = DecisionTreeRegressor(random_state=42)
model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(x_test)
```

```python
# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate the R² score
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

So, for this dataset after applying model we have more error therefore prediction can vary.

```python
# Make predictions (lets take new data n predict the house price)
new_data = pd.DataFrame({
    'bedrooms': [2],
    'bathrooms': [2],
    'sqft_living': [1800],
    'sqft_lot': [6000],
    'floors': [3],
    'waterfront': [0],
    'view': [0],
    'grade': [7],
    'yr_built': [2000],
    'yr_renovated': [0],
    'zipcode': [98103],
    'lat': [47.6608],
    'long': [-122.291],
    'sqft_living15': [1600],
    'sqft_lot15': [6000]
})

predicted_price = model.predict(new_data)

print("Predicted Price:", predicted_price)
```
